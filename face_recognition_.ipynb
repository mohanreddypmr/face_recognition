# Face Detection:

```python
image=cv2.imread('/content/images111.jpeg')

image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
face_locations=face_recognition.face_locations(image)

number_of_faces=len(face_locations)

print('input image contains '+str(number_of_faces)+' faces')
```

```python
plt.imshow(image)

ax=plt.gca()

for face_location in face_locations:
    top,right,bottom,left=face_location
    x,y,w,h=left,top,right,bottom

    rect=Rectangle((x,y),w-x,h-y,fill=False,color='red')
    ax.add_patch(rect)

plt.show()
```

# Face Recognition using face_recognition library:

```python
image=cv2.imread('/content/arjun.jpg')

arjun= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

image=cv2.imread('/content/prabhas.jpeg')

prabhas= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

image=cv2.imread('/content/nani.jpeg')

nani= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
arjun_encoding=face_recognition.face_encodings(arjun)[0]

prabhas_encoding=face_recognition.face_encodings(prabhas)[0]

nani_encoding=face_recognition.face_encodings(nani)[0]
```

```python
plt.title('Arjun photo')
plt.imshow(arjun)
```

```python
plt.title('prabhas photo')
plt.imshow(prabhas)
```

```python
plt.title('nani photo')
plt.imshow(nani)
```

```python
known_faces=[arjun_encoding,prabhas_encoding,nani_encoding]
```

```python
image=cv2.imread('/content/nani6.jpg')


unknown_image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(unknown_image)

unknown_face_encodings=face_recognition.face_encodings(unknown_image)
```

Recognize the face:

```python
from scipy.spatial import distance

for unknown_face_encoding in unknown_face_encodings:
    result=[]
    for known_face in known_faces:
        d=distance.euclidean(known_face,unknown_face_encoding)
        result.append(d)
    threshold=0.5
    result=np.array(result)<=threshold

    name='unknown'

    if result[0]:
        name='arjun'
    elif result[1]:
        name='prabhas'
    elif result[2]:
        name='nani'
    print('found '+name+' in the image')
```
